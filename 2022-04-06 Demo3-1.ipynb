```python
%pylab inline
from sci378 import *
from lmfit import *
from pyndamics3 import Simulation
from pyndamics3.fit import fit, Parameter
```

```python
data=pd.read_csv('fludata3.csv')
data=data[data['region']=='Bay Area']
data=data[data['season']=='2018-2019']
data
```

```python
t_data=data['weekending']
t_data=date_to_float(data['weekending'])
t_data=t_data-min(t_data)
y_data=data['Percent_ILI']
```

```python
#SEIR
sim=Simulation()
sim.add("N=S+E+I+R",1000)
sim.add("S'=-*S*I/N",6)
sim.add("E'= *I*S/N-*E",.5)
sim.add("I'=*E-*I",2,plot=True)
sim.add("R'=*I",1)
sim.add_data(t=t_data,I=y_data,plot=True)
sim.params(=2,=10,=.5)
sim.run(1)
```

```python
result1=fit(sim,
           Parameter("",value=1,min=0),
           Parameter("",value=1,min=0),
           Parameter("",value=1,min=0))
result1
```

```python
plot(t_data,y_data,'ko')
plot(sim.t,sim.I)
title('SEIR Flu SF18')
xlabel('Time')
ylabel('Percent Infected')
```

```python
#SIR
simm=Simulation()
simm.add("N=S+I+R",1000)
simm.add("S'=-*S*I/N",6)
simm.add("I'=*S*I/N-*I",2,plot=True)
simm.add("R'=*I",.5)
simm.add_data(t=t_data,I=y_data,plot=True)
simm.params(=2,=10,=.5)
simm.run(1)
```

```python
result2=fit(simm,
           Parameter("",value=1,min=0),
           Parameter("",value=1,min=0),
           Parameter("",value=1,min=0))
result2
```

```python
plot(t_data,y_data,'ko')
plot(simm.t,simm.I)
title('SIR Flu SF18')
xlabel('Time')
ylabel('Percent Infected')
```

```python
plot(t_data,y_data,'-o',color='#CD3828')
plot(sim.t,sim.I,'-',color='#2EA3CE')
plot(simm.t,simm.I,'g-')
title("SIR vs SEIR on Influenza, SF 18-19")
xlabel("time")
ylabel("percent infected")
```
