```python
#3.7.4 Ndot=rN(1-N/K)-H(N/A+N)
#H>0, A>0, N=0 is fixed pt, N nd Ndot are variables (x and y)

Ndot = 1
N= 1
dn = .1

#parameters
H = .5
A = 4
K=2
r = 3

S=Storage()
S+=N,Ndot

while N<20:
    dNdot = r*N*((1-N)/K)-(H*(N/(A+N)))
    
    N=N+dn
    Ndot=Ndot+dNdot
    
    S+=N,Ndot
N,Ndot=S.arrays()
plot(N,Ndot)
title('3.7.4: Refined Fishery Model')
xlabel('population?')
ylabel('fish catching rate')
#keep the hex in lowercase
#is h a variable instead of a parameter?
```

```python
for H in [.5,20,80]:
    Ndot = 1
    N= 1
    dn = .1

    #parameters
    A = 4
    K=2
    r = 3

    S=Storage()
    S+=N,Ndot

    while N<20:
        dNdot = r*N*((1-N)/K)-(H*(N/(A+N)))

        N=N+dn
        Ndot=Ndot+dNdot

        S+=N,Ndot
    N,Ndot=S.arrays()
    plot(N,Ndot, label=f"H= {H}")
    title('3.7.4: Refined Fishery Model')
    xlabel('x')
    ylabel('fish catching rate')
legend()
```
