```python
%pylab inline
from pyndamics3 import Simulation
from sci378 import *
```

```python
data=pd.read_csv('fludata.csv')
data=data[data['region']=='Bay Area']
data=data[data['season']=='2018-2019']
data
```

```python
t_data=data['weekending']
t_data=date_to_float(data['weekending'])
t_data=t_data-min(t_data)
y_data=data['Percent_ILI']

plot(t_data,y_data,'o')
```

```python
sim=Simulation()
sim.add("=*(*I+m)",1,plot=True)
sim.add("S'=-*S",1,plot=True)
sim.add("I'=*S-*I",1,plot=True)
sim.add("R'=*I",1,plot=True)
sim.params(=2,=.3,m=1,=1)
#sim.add_data(t=t_data,x=y_data,plot=True)
sim.run(1)
```

```python
t_data=data['weekending']
t_data=date_to_float(data['weekending'])
t_data=t_data-min(t_data)
y_data=data['Percent_ILI']

sim=Simulation()
sim.add("=*(*I+m)")
sim.add("S'=-*S",5.75)
sim.add("I'=*S-*I",.7)
sim.add("R'=*I",.5)
sim.params(=14,=.25,m=.005,=1.5)
#sim.add_data(t=t_data,y=y_data,plot=True)
sim.run(1)
set_plot_colors('summer',3)
plot(sim.t,sim.I,)
plot(t_data,y_data,'o-')
title('Influenza Transmission SIR Model in the Bay Area, 2018')
xlabel('Time')
ylabel('Percent Infected')
```
